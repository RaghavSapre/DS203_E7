<h3>Data Preparation</h3>

```python
Main_data = pd.read_csv("e7-htr-currernt.csv")
```

Adding Day, Time to the data

```python
def convert_time_format(time_str):
    hours, minutes = map(int, time_str.split(':'))
    
    formatted_time = '{:02d}:{:02d}:00'.format(hours, minutes)
    
    return formatted_time

day_calc = Main_data['Timestamp']
Day = [day_calc[i][:2] for i in range(0, len(day_calc))]

Main_data['Day'] = Day

time_calc = Main_data['Timestamp']
Time = [convert_time_format(time_calc[i][12:18]) for i in range(0, len(time_calc))]

Main_data['Time'] = Time

Main_data
```
